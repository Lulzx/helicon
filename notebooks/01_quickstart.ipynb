# Helicon Quick Start
This notebook demonstrates the core Helicon workflow: compute a magnetic field, inspect the field topology, and run a quick validation check.

```python
import numpy as np
import helicon as mnx
from helicon.fields.biot_savart import Coil, Grid, compute_bfield
print(f'Helicon version: {mnx.__version__}')
```

## 1. Define coil geometry
A single solenoid coil at z=0, r=0.1 m with 10 kA-turns.

```python
coils = [Coil(z=0.0, r=0.1, I=10000.0)]
grid = Grid(z_min=-0.5, z_max=2.0, r_max=0.5, nz=128, nr=64)
bfield = compute_bfield(coils, grid)
print(f'B_z max (on-axis): {bfield.Bz.max():.4f} T')
print(f'Grid shape: {bfield.Bz.shape}')
```

## 2. Analytical pre-screening
Use the Tier 1 analytical model to quickly estimate nozzle performance.

```python
from helicon.optimize.analytical import screen_geometry
result = screen_geometry(coils, z_min=-0.5, z_max=2.0)
print(f'Mirror ratio R_B = {result.mirror_ratio:.2f}')
print(f'Thrust efficiency \u03b7_T = {result.thrust_efficiency:.3f}')
print(f'Plume half-angle \u03b8 = {result.divergence_half_angle_deg:.1f}\u00b0')
print(f'Thrust coefficient C_T = {result.thrust_coefficient:.3f}')
```

## 3. Configuration from YAML
Load a preset engine configuration.

```python
config = mnx.config.SimConfig.from_preset('dfd')
print(f'Nozzle type: {config.nozzle.type}')
print(f'Species: {config.plasma.species}')
print(f'T_i = {config.plasma.T_i_eV} eV, T_e = {config.plasma.T_e_eV} eV')
```
